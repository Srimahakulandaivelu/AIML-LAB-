**TASK 2:COVARIANCE AND CORRELATION**

The covariation matrix measures how two variables change together and the correlation matrix measures both the strength and direction of  a linear relationship between two variables.

```python
import pandas as pd
df = pd.read_csv('Titanic-Dataset.csv')

numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()
print(correlation_matrix)
covariance_matrix = numerical_df.cov()
print(covariance_matrix)
```

**INFERENCE**:Here the fare column has a strong negative correlation -0.549 as expected because the higher class passengers paid more for their tickets.We could also see that the older passengers tend to be in the higher class by the moderate negative correlation in the age column -0.369.

The correlation matrix when overviewed on the basis of the survival shows a positive covariance of 6.221 in the fare column giving the inference that those who paid more where more likely to survive.

**TASK 3:QUANTILES AND RANGES**

Quantiles divide our dataset into equal sized groups.The ranges the difference between the maximum and minimum values of the dataset

```python
import pandas as pd
df = pd.read_csv('Titanic-Dataset.csv')

numerical_df = df.select_dtypes(include=['number'])
quantiles = numerical_df.quantile([0.25, 0.5, 0.75])
print(quantiles)


for column in numerical_df.columns:
    col_range = numerical_df[column].max() - numerical_df[column].min()
    print(f"Range of {column}: {col_range}")
```

**INFERENCE:**
Survival Rate: More than half of the passengers in the dataset did not survive.

Class Distribution: The majority of passengers were in 3rd class.

Age: Passengers tended to be relatively young, with a median age of 28.

Family: Most passengers were traveling alone, with no siblings/spouses or parents/children aboard.

Fare: Most of the passengers paid a low fare.

**SCATTER PLOT:**It is a type of graph that displays the relationship between two continuous variables.

```python
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Titanic-Dataset.csv")

# Scatter plot using matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(df['PassengerId'], df['Age'], alpha=0.5)
plt.title('Scatter Plot of Age vs Passenger ID',fontweight='bold',fontsize='18')
plt.xlabel('Passenger ID',fontweight='bold',fontsize='18')
plt.ylabel('Age',fontweight='bold',fontsize='18')
plt.show()
```

**INFERENCE**:Here we have used passenger id as x axis and age as y axis.we could see that more no of passengers are present in the age group 20-40.

**PIE** **CHART**:It is a specialised chart designed specifically for how a whole is divided into different parts based on their categories and distribution

```python
import matplotlib.pyplot as plt

# Count the number of males and females
gender_counts = df['Sex'].value_counts()

# Create a pie chart for gender distribution
plt.figure(figsize=(10, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['blue', 'red'])

# Adding title
plt.title('Gender Distribution',fontweight='bold',fontsize='18')

# Show plot
plt.show()
```

**INFERENCE**:Here the pie chart is classified as male and female and we could see that the no of male passenger is more with a percentage of 64.8 meanwhile the female passengers is of the percentage 35.2
