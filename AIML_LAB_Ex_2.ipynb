**BREADTH FIRST SEARCH ALGORITHM (WITH VISUALIZATION)**

```python
from collections import deque
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

Libraries needed are imported at the starting of the program

```python
def bfs (maze,start,goal):
  queue = deque([(start,[])])
  visited = set()

  while queue:
    current, path = queue.popleft()
    x, y = current

    if current == goal:
      return path + [current]

    if current in visited:
      continue

    visited.add(current)

    for dx, dy in [(0,1), (0,-1), (1,0), (-1,0)]:
      nx , ny = x + dx, y + dy

      if 0 <= nx < len(maze) and 0 <= ny < len(maze[0]) and maze[nx][ny] != 1:
        queue.append(((nx,ny), path + [current]))

  return None
```

  defining a function bfs to calculate the bfs of a matrix(maze)

  queue : a tuple initialised with start position and an empty path

  visited : a set to keep track of the visited points
  
  As long as the queue is not empty, the loop continues
  
  the current, path = queue.popleft() is that the current position and the path is the first element in the queue.
  
  If the goal is reached , it is returned along with its previous path
  
  If te current position is already in the path, just continue the loop
  
  If it is just visited, add it to the visited set
  
  if the dx, dy co-ord offsets are in the given array it makes moves based on these co-ords

  (0, 1): Move right
  (0, -1): Move left
  (1, 0): Move down
  (-1, 0): Move up

  
  After making the moves, updating the new co-ords nx, ny.

  
  then checking if the nx and ny are within the maze limits and then skipping the obstacles, adding it to the queue(path)

  If the goal is not reachable, returning none

```python
from random import choice
```

this library is used to select a random value in the given range

```python
combo_indices=[
    [0,1,2],
    [3,4,5],
    [6,7,8],
    [0,3,6],
    [1,4,7],
    [2,5,8],
    [0,4,8],
    [2,4,6]
]

empty_sign = '.'
ai_sign = 'X'
opponent_sign = 'O'
```

combo_indices = defines all the possible combinations for the 3*3 grid

empty_sign = symbol for empty sign (.)

ai_sign = symbol for AI player (X)

opponent_sign = symbol for opponent player (O)

```python
def print_board(board):
  print(" ")
  print(''.join(board[:3])) #first row
  print(''.join(board[3:6])) #second row
  print(''.join(board[:6])) #last row
  print(" ")
```

print_board function is used to display the rows of the game,

```python
def opponent_move(board, row, column):
  index = 3 * (row-1) + (column-1)
  # this index calculation is used to convert the row and column based indices(1-3) to zero based indices(0-9)
  # Eg : index = 3 * (2-1) + (3-1) = 3 + 2 = 5

  if board[index] == empty_sign:
  # checking if the target cell is empty or not

    board[index] = opponent_sign
    # updating the game board index to the opponent's sign

  return board
  # returning or displaying the board after each updation
```

```python
def all_moves_from_board(board,sign):
# defining a function named all_moves_from_board that takes the board and sign as input

  move_list=[]
  # a list to store all the possible new board states after the player makes a valid move

  for i,v in enumerate(board):
  # iterating over the board
  # Returns both the index (i) and the value (v) of each element in the board list.
  # i = to locate the empty spots on the board
  # v = the current value at the index

    if v == empty_sign: # if the current value is empty
      move_list.append(board[:i] + sign + board[i+1:])

      # board[:i]: Gets all elements before the empty spot.
      # sign: The player's sign ('X' or 'O'), which fills the empty spot.
      # board[i+1:]: Gets all elements after the empty spot.
      # Combine all three to create a new board state.

  return move_list
```

```python
def ai_move(board):
  # defining a function ai_move(board) that returns a random value to fill in out of the possible choices

  return choice(all_moves_from_board(board, ai_sign))
```

**8 QUEENS PROBLEM USING DEPTH FIRST SEARCH**

```python
class NQueens:
  def __init__(self,size):
    self.size = size
    self.solutions = 0
    self.solve()
  def solve(self):
    positions = [-1] * self.size
    self.put_queen(positions,0)
    print("Found", self.solutions, "solutions.")
  def put_queen(self, positions, target_row):
    if target_row == self.size:
      self.show_full_board(positions)
      self.solutions += 1
    else:
      for column in range(self.size):
        if self.check_place(positions, target_row, column):
          positions[target_row] = column
          self.put_queen(positions, target_row+1)
  def check_place(self, positions, occupied_rows, column):
    for i in range(occupied_rows):
      if positions[i] == column or positions[i] - i == column - occupied_rows or positions[i] + i == column + occupied_rows:
        return False
    return True
  def show_full_board(self, positions):
    for row in range(self.size):
      line  = ""
      for column in range(self.size):
        if positions[row] == column:
          line += "Q"
        else:
          line += "."
      print(line)
    print("/n")
```
